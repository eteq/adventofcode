```julia
partsum = 0

for j in 1:length(input)
    line = input[j]
    for digmatch in eachmatch(digreg, line, overlap=true)
        
        #ledge = digmatch.start > 1 ? (digmatch.start - 1) : 1
        #redge = digmatch.stop == length(line) ? digmatch.stop : (digmatch.stop + 1)
        partnum = parse(Int, digmatch.captures[2])
        
        ledge = digmatch.offsets[1]
        redge = digmatch.offsets[3]
        if redge > length(line)
            redge = length(line)
        end
        
        adj = (symbol_map[ledge, j] != 0) || (symbol_map[redge, j] != 0)
        if (!adj) && (j > 1)
            # check the top
            for i in ledge:redge
                if symbol_map[i, j-1] != 0
                    adj = true
                    break
                end
            end
        end
        
        if (!adj) && (j < length(input))
            # check the bottom
            for i in ledge:redge
                if symbol_map[i, j+1] != 0
                    adj = true
                    break
                end
            end
        end
        if adj
            partsum += partnum
        end
    end
end

partsum
```
