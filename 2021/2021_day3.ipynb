```rust
fn compute_gamma(input: &str) -> (isize, isize) { 
    let nbits = input.lines().next().unwrap_or("").chars().count();
    
    let mut onecounts = vec![0; nbits];
    
    let mut n = 0;
    for line in input.lines() {
        n += 1;
        for (i, chr) in line.chars().enumerate() {
            if chr == '1' { onecounts[i] += 1}
        }
    }
    
    let mut gammaval = 0;
    let mut epsval = 0;
    for (i, onecount) in onecounts.iter().enumerate() {
        if *onecount >= n-onecount {
            gammaval += 1 << (nbits-1-i);
        } else {
            epsval += 1 << (nbits-1-i); 
        }
    }
    (gammaval,epsval)
}

compute_gamma(test_input)
```

```rust
enum LifeSupportKind { O2, CO2 }

fn compute_lifesupport(input: &str, kind: LifeSupportKind) -> isize { 
    let nbits = input.lines().next().unwrap_or("").chars().count();
    let nlines = input.lines().count();
    
    let mut includeline = vec![true; nlines];
    let mut answer = 0;
    
    for bit in 0..nbits {
        let mut ones = 0;
        let mut zeros = 0;
        for (i, line) in input.lines().enumerate() {
            if includeline[i] {
                match line.as_bytes().get(bit) {
                    Some(48) => {zeros += 1;},
                    Some(49) => {ones += 1;},
                    _ => {println!("not a 1 or a 0...", );},
                }
            }
        }
        
        let goodbyte = match kind {
            LifeSupportKind::O2 => if ones >= zeros { 49 } else { 48 },
            LifeSupportKind::CO2 => if ones < zeros { 49 } else { 48 },
        };
        
        let mut countlines = 0;
        for (i, line) in input.lines().enumerate() {
            if includeline[i] {
                if line.as_bytes().get(bit).unwrap() != &goodbyte {
                    includeline[i] = false;
                } else {
                    countlines +=1 ;
                }
            }
        }
        if countlines <= 1 {
            for (i, line) in input.lines().enumerate() {
                if includeline[i] {
                    // this is the magic one
                    for j in  0..nbits {
                        if *line.as_bytes().get(j).unwrap() == 49 {
                            answer += 1 << nbits-1-j 
                        }                        
                    }
                }
            }
        }
        if answer > 0 { break; }
    }
    answer
}

compute_lifesupport(test_input, LifeSupportKind::O2) * compute_lifesupport(test_input, LifeSupportKind::CO2)
```
