```rust
// 0 indicates no error
fn syntax_error_for_line(line :&str) -> usize {
    let mut braces: Vec<char> = Vec::new();
    for chr in line.chars() {
        match chr {
            '(' | '{' | '<' | '[' => { braces.push(chr); },
            ')' => { if braces.pop().unwrap_or(')') != '(' { return 3 }},
            ']' => { if braces.pop().unwrap_or('[') != '[' { return 57 }},
            '}' => { if braces.pop().unwrap_or('{') != '{' { return 1197 }},
            '>' => { if braces.pop().unwrap_or('<') != '<' { return 25137 }},
            _ => { println!("invalid char!: {}", chr); }
        }
    }
    0
}


let mut score = 0;
for line in test_input.lines() { score += syntax_error_for_line(line); }
score
```

```rust
// 0 indicates no error
fn line_completion_score(line :&str) -> usize {
    let mut braces: Vec<char> = Vec::new();
    for chr in line.chars() {
        match chr {
            '(' | '{' | '<' | '[' => { braces.push(chr); },
            ')' => { if braces.pop().unwrap_or(')') != '(' { return 0 }},
            ']' => { if braces.pop().unwrap_or('[') != '[' { return 0 }},
            '}' => { if braces.pop().unwrap_or('{') != '{' { return 0 }},
            '>' => { if braces.pop().unwrap_or('<') != '<' { return 0 }},
            _ => { println!("invalid char!: {}", chr); }
        }
    }
    
    // needs to be completed
    let mut score = 0;
    
    while braces.len() > 0 {
        let val = braces.pop().unwrap();
        score *= 5;
        match val {
            '(' => { score += 1; },
            '[' => { score += 2; },
            '{' => { score += 3; },
            '<' => { score += 4; },
            _ => {}
        }
    }
    score
}

let mut sortvec: Vec<usize> = Vec::new(); 
for line in test_input.lines() { 
    let score = line_completion_score(line);
    if score > 0 { sortvec.push(score); }
}
sortvec.sort();
sortvec[sortvec.len()/2]
```
