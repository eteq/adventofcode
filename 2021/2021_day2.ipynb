```rust
fn follow_course(input: &str) -> (isize, isize) {
    let mut x = 0;
    let mut z = 0;
    
    for line in input.lines() {
        let mut iter = line.split_ascii_whitespace();
        let direction = iter.next().unwrap();
        let distance : isize = iter.next().unwrap().parse().unwrap();
        match direction {
            "forward" => x += distance,
            "up" => z += distance,
            "down" => z -= distance,
            _ => println!("Found an unexpected linedirection! {}", line),
        }
    }
    (x, -z)
}
follow_course(test_input)
```

```rust
fn follow_course2(input: &str) -> (isize, isize) {
    let mut x = 0;
    let mut z = 0;
    let mut aim = 0;
    
    for line in input.lines() {
        let mut iter = line.split_ascii_whitespace();
        let direction = iter.next().unwrap();
        let distance : isize = iter.next().unwrap().parse().unwrap();
        match direction {
            "forward" => {x += distance; z+= aim*distance},
            "up" => aim -= distance,
            "down" => aim += distance,
            _ => println!("Found an unexpected linedirection! {}", line),
        }
    }
    (x, z)
}
follow_course2(test_input)
```
