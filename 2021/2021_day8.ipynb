````rust
fn input_to_vecs(input: &str) -> (Vec<String>, Vec<String>) {
    let flat_string = str::replace(input, "|\n", "| ");
    let mut pair = flat_string.split(" | ");
    let v1 = pair.next().unwrap_or("").split(" ").map(String::from).collect();
    let v2 = pair.next().unwrap_or("").split(" ").map(String::from).collect();
    
    (v1, v2)
} 
input_to_vecs(test_input)
````

````rust
fn count_uniques(v: Vec<String>) -> usize {
    let mut uniquecount = 0;
    
    for elem in v {
        match elem.len() {
            2 => { uniquecount+= 1; } , //digit 1
            4 => { uniquecount+= 1; } , //digit 4
            3 => { uniquecount+= 1; } , //digit 7
            7 => { uniquecount+= 1; } , //digit 8
            n => { } ,  // no op
        }
    }
    uniquecount
}

// should be 4 from the example
count_uniques(input_to_vecs("fgeab ca afcebg bdacfeg cfaedg gcfdb baec bfadeg bafgc acf |
gebdcfa ecba ca fadegcb").1)
````

````rust
use std::time::Instant;
use std::collections::HashMap;

const SEGMENTS: &str = "abcdefg";

fn digit_vec_to_map(digits: Vec<String>) -> HashMap<char,usize> {
    let mut char_to_segment_index: HashMap<char,usize> = HashMap::new();
    
    let mut len2_idx: usize = 100;
    let mut len3_idx: usize = 100;
    let mut len4_idx: usize = 100;
    let mut len7_idx: usize = 100;
    let mut len5_idxs: Vec<usize> = Vec::new();
    let mut len6_idxs: Vec<usize> = Vec::new();
    
    let mut chrf: char = '0';
    
    for (i, s) in digits.iter().enumerate() {
        match s.len() {
            2=> {len2_idx = i;},
            3=> {len3_idx = i;},
            4=> {len4_idx = i;},
            5=> {len5_idxs.push(i);},
            6=> {len6_idxs.push(i);},
            7=> {len7_idx = i;},
            d=> {println!("IMPOSSIBLE LENGTH! {}", d)},
        }
    }
    // a -> a appears only in #7 and #8 but not #1 and #4 
    for chr in SEGMENTS.chars() {
        if digits[len3_idx].find(chr) != None && 
           digits[len7_idx].find(chr) != None &&
           digits[len2_idx].find(chr) == None &&
           digits[len4_idx].find(chr) == None { 
               char_to_segment_index.insert(chr, 0); 
               break; 
        }
    }
    // f -> f appears in all digits except for one of the len-5 digits
    'fsearch: for chr in SEGMENTS.chars() {
        if digits[len2_idx].find(chr) == None { continue; }
        if digits[len3_idx].find(chr) == None { continue; }
        if digits[len4_idx].find(chr) == None { continue; }
        if digits[len7_idx].find(chr) == None { continue; }
        for i in len6_idxs.iter() { if digits[*i].find(chr) == None { continue 'fsearch; } }
        
        let mut count5 = 0;
        for i in len5_idxs.iter() { if digits[*i].find(chr) != None { count5 += 1; } }
        if count5 == 2 { 
            char_to_segment_index.insert(chr, 5); 
            chrf = chr;}
    }
    // c -> c and f appear in all of the unique-digit ones
    for chr in SEGMENTS.chars() {
        if chr == chrf { continue; }
        if digits[len3_idx].find(chr) != None && 
           digits[len7_idx].find(chr) != None &&
           digits[len2_idx].find(chr) != None &&
           digits[len4_idx].find(chr) != None { 
               char_to_segment_index.insert(chr, 2); 
               break; 
        }
    }
    // d, g
    // the length 5 that has a,c,f is #3, so it's remaining 2 are d and g. 
    // d is the one that'd in #4/len-4, and g is the one that's not
    for i in len5_idxs.iter() {
        let mut matchcount = 0;
        for k in char_to_segment_index.keys() { if digits[*i].find(*k) != None { matchcount+=1; }}
        if matchcount != char_to_segment_index.len() { continue; }
        
        for chr in digits[*i].chars() {
            if ! char_to_segment_index.contains_key(&chr) {
                if digits[len4_idx].find(chr) == None { 
                    char_to_segment_index.insert(chr, 6); 
                } else {
                    char_to_segment_index.insert(chr, 3); 
                }
                
            }
        }
            
    }
    //  b & e -> of the remaining two, b is in #4/len-4, but e is not
    let donekeys: Vec<char> = char_to_segment_index.keys().cloned().collect();
    for chr in SEGMENTS.chars() {
        if ! donekeys.contains(&chr) {
            if digits[len4_idx].find(chr) == None {
                char_to_segment_index.insert(chr, 4); 
            } else {
                char_to_segment_index.insert(chr, 1); 
            }
        }
    }
    
    char_to_segment_index
}

let now = Instant::now();
let res = digit_vec_to_map(input_to_vecs(test_input).0);
let el = now.elapsed();
// time estimate of one run and the  full run
(res, el, el*input8.lines().count() as u32) 
````

````rust
fn outstr_to_digit(outstr:&str, map:&HashMap<char,usize>) -> u8 {
    
    // really should have this be saved globally but eh its fast enough
    let mut code_to_number = HashMap::new();
    // the encoding here is just binary by letter code
    code_to_number.insert((1 << 0) + (1 << 1) + (1 << 2)+ (1 << 4) + (1 << 5) + (1 << 6), 0);
    code_to_number.insert((1 << 2) + (1 << 5), 1);
    code_to_number.insert((1 << 0) + (1 << 2) + (1 << 3) + (1 << 4) + (1 << 6), 2);
    code_to_number.insert((1 << 0) + (1 << 2) + (1 << 3) + (1 << 5) + (1 << 6), 3);
    code_to_number.insert((1 << 1) + (1 << 2) + (1 << 3) + (1 << 5), 4);
    code_to_number.insert((1 << 0) + (1 << 1) + (1 << 3) + (1 << 5) + (1 << 6), 5);
    code_to_number.insert((1 << 0) + (1 << 1) + (1 << 3) + (1 << 4) + (1 << 5) + (1 << 6), 6);
    code_to_number.insert((1 << 0) + (1 << 2) + (1 << 5), 7);
    code_to_number.insert((1 << 0) + (1 << 1) + (1 << 2) + (1 << 3) + (1 << 4) + (1 << 5) + (1 << 6), 8);
    code_to_number.insert((1 << 0) + (1 << 1) + (1 << 2) + (1 << 3) + (1 << 5) + (1 << 6), 9);
    
    let mut codeval = 0;
    for chr in outstr.chars() { 
        codeval += 1 << map[&chr];
    }
    let digit = code_to_number.get(&codeval);
    match digit {
        Some(v) => { return *v as u8; }
        None => { 
            println!("invalid code {}", codeval);
            return 255; },
    }
}

let map = digit_vec_to_map(input_to_vecs(test_input).0);
assert_eq!(outstr_to_digit("cdfeb", &map), 5);
assert_eq!(outstr_to_digit("fcadb", &map), 3);
assert_eq!(outstr_to_digit("cdfeb", &map), 5);
assert_eq!(outstr_to_digit("cdbaf", &map), 3);
````

````rust
fn line_to_output(line: &str) -> usize {
    let (udigits, outputs) = input_to_vecs(line);
    let map = digit_vec_to_map(udigits);
    
    let mut result = 0;
    for (i, digstr) in outputs.iter().rev().enumerate() {
        let dig = outstr_to_digit(digstr, &map);
        result += dig as usize * 10_usize.pow(i as u32);
    }
    result
}
assert_eq!(line_to_output(test_input), 5353);
````
